## Plot curves from several wells

```python
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
alias = {'GR': ['GR', 'GRC', 'NGT']}
```

```python
fig, axs = plt.subplots(figsize=(7, 14), ncols=2)

for i, w in enumerate(p):
    ax = axs[i]
    gr = w.get_curve('GR', alias=alias)
    ax = gr.plot(ax=ax)
    ax.set_title("GR for {}".format(w.uwi))
    
plt.show()
```
